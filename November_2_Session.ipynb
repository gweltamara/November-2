#**GITHUB TERMINOLOGIES**

**Version Control:**

  Imagine a magical time-travel system for your files. It helps you keep track of changes you make to your documents over time so you can go back to any previous version if something goes wrong.

**Branch:**

  Think of it as a separate copy of your project where you can work on new ideas or features without messing up the main project. It's like having a different sandbox to play in.

**Commit:**

When you're happy with the changes you've made in your project, you "save" them in a way that records what you did. It's like taking a snapshot of your work at a specific point in time.

**Repository:**

  This is like a big digital folder or storage space for all your project files. It holds everything you need for your project, including its history and different branches.

**SHA (Secure Hash Algorithm):**

  It's like a unique fingerprint for every commit in your project. This fingerprint ensures that each snapshot is distinct and can be identified, making it easy to find and work with specific versions of your project.
